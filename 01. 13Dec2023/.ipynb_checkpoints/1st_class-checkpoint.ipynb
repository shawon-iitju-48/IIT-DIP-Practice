Read original image


```python
import cv2
print("original image")
image=cv2.imread("pic1.jpg")
from google.colab.patches import cv2_imshow
cv2_imshow(image)
```

Gray scale image

```python
print("gray scale image")
gray=cv2.imread('pic1.jpg',0)
cv2_imshow(gray)
cv2.imwrite('pic1.jpg',gray)
print(image)
```

Image rotation


```python
import cv2
import imutils
image=cv2.imread("pic1.jpg")
from google.colab.patches import cv2_imshow
cv2_imshow(image)
print("rotated image")
print("45 degree rotation")
r1=imutils.rotate(image,45)
cv2_imshow(r1)
```

CMI image


```python
import cv2
print("original image")
image=cv2.imread("pic1.jpg")
from google.colab.patches import cv2_imshow
cv2_imshow(image)
print("RGB image")
R,G,B=cv2.split(image)
print('R')
cv2_imshow(R)
print('G')
cv2_imshow(G)
print('B')
cv2_imshow(B)
C=1-R
M=1-G
Y=1-B
print("C image")
cv2_imshow(C)
print("M image")
cv2_imshow(M)
print("Y image")
cv2_imshow(Y)

```
