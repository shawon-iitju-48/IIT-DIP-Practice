# Image importing and flipping

```python
from PIL import Image
original_img=Image.open("ironman.png")
vertical_img=original_img.transpose(method=Image.FLIP_TOP_BOTTOM)
vertical_img.save("vertical_ironman.png")
```

```python
original_img.show()
```

```python
vertical_img.size
```

# Resizing

```python
size=(50,50)
new_img=vertical_img.resize(size)
```

```python
new_img.save("new_resized_vertical_img.png")
```

```python
new_img
```

# Cropping image

```python
img=original_img
width,height=img.size

left=100
top=height/4
right=300
bottom=3*height/4
iml=img.crop((left,top,right,bottom))
iml
```

```python
import os
from PIL import Image
from matplotlib import pyplot as plt
root="Images"
fnames=os.listdir(root)

fig,axs=plt.subplots(nrows=1,ncols=len(fnames),figsize=(20,20))
axs=axs.flatten()

for i in range(3):
    filepath=os.path.join(root,fnames[i])
    img=Image.open(filepath)
    axs[i].imshow(img)
    axs[i].axis('off')
    axs[i].set_title(fnames[i])
plt.show()
```

# Convert OpenCV image to PIL image in Python

```python
import cv2

opencv_img=cv2.imread("ironman.png")

color_converted=cv2.cvtColor(opencv_img,cv2.COLOR_BGR2RGB)

#cv2.imshow("OpenCV Image", opencv_img)

pil_image=Image.fromarray(color_converted)
pil_image.show()
```

```python
plt.imshow(color_converted)
```
