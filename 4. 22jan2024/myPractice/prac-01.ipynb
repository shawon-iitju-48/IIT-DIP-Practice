```python
import os
root="Images"
fnames=os.listdir(root)
```

```python
coln=len(fnames)
```

```python
fig,axs=plt.subplots(nrows=1,ncols=coln,figsize=(10,10))

for i in range(coln):
    filepath=os.path.join(root,fnames[i])
    timg=Image.open(filepath)
    axs[i].imshow(timg)
    axs[i].axis("off")
    axs[i].set_title(fnames[i])
plt.show()
```
