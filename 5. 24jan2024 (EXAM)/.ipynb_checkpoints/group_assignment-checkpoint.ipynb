```python
import cv2
import numpy as np

def apply_arithmetic_mean_filter(image):
    height, width = image.shape[:2]
    result = np.zeros((height, width), dtype=np.uint8)

    # Iterate over each pixel in the image (excluding border pixels)
    for i in range(1, height - 1):
        for j in range(1, width - 1):
            # Compute the average of the 3x3 neighborhood
            average = np.mean(image[i-1:i+2, j-1:j+2])

            # Assign the average value to the corresponding pixel in the result image
            result[i, j] = np.round(average).astype(np.uint8)

    return result

# Load your black and white image
image_path = "black_white_1st_pic.png"  # Replace with your image path
input_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded successfully
if input_image is None:
    print("Error: Image not loaded.")
else:
    # Apply the arithmetic mean filter
    output_image = apply_arithmetic_mean_filter(input_image)

    # Display the original and filtered images
    cv2.imshow("Original Image", input_image)
    cv2.imshow("Arithmetic Mean Filter", output_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_arithmetic_mean_filter(image, kernel_size):
    height, width = image.shape[:2]
    result = np.zeros((height, width), dtype=np.uint8)
    padding = kernel_size // 2

    for i in range(padding, height - padding):
        for j in range(padding, width - padding):
            average = np.mean(image[i-padding:i+padding+1, j-padding:j+padding+1])
            result[i, j] = np.round(average).astype(np.uint8)
    return result

def plot_results(original, filtered_list, titles):
    plt.figure(figsize=(12, 4))
    # Plot the original image
    plt.subplot(1, len(filtered_list) + 1, 1)
    plt.imshow(original, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    # Plot the filtered images
    for i in range(len(filtered_list)):
        plt.subplot(1, len(filtered_list) + 1, i + 2)
        plt.imshow(filtered_list[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')

    plt.show()

image_path = "black_white_1st_pic.png" 
input_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if input_image is None:
    print("Error: Image not loaded.")
else:
    filter_sizes = [3, 5, 7]
    filtered_images = [apply_arithmetic_mean_filter(input_image, size) for size in filter_sizes]
    plot_titles = [f'{size}x{size}' for size in filter_sizes]
    plot_results(input_image, filtered_images, plot_titles)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_arithmetic_mean_filter(image, kernel_size):
    height, width = image.size[:2]
    result = np.zeros((height, width), dtype=np.uint8)
    padding = kernel_size // 2
    image=np.array(image)
    for i in range(padding, height - padding):
        for j in range(padding, width - padding):
            average = np.mean(image[i-padding:i+padding+1, j-padding:j+padding+1])
            result[i, j] = np.round(average).astype(np.uint8)
    return result

def plot_results(original, filtered_list, titles):
    plt.figure(figsize=(12, 4))
    # Plot the original image
    plt.subplot(1, len(filtered_list) + 1, 1)
    plt.imshow(original, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')
    # Plot the filtered images
    for i in range(len(filtered_list)):
        plt.subplot(1, len(filtered_list) + 1, i + 2)
        plt.imshow(filtered_list[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')

    plt.show()

#image_path = "cow_img.jpg"
from PIL import Image
img = Image.open('cow_img.jpg')
thresh = 200
fn = lambda x : 255 if x > thresh else 0
black_white_img = img.convert('L').point(fn, mode='1')


#input_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
input_image=black_white_img
if input_image is None:
    print("Error: Image not loaded.")
else:
    filter_sizes = [3, 5, 7]
    filtered_images = [apply_arithmetic_mean_filter(input_image, size) for size in filter_sizes]
    plot_titles = [f'{size}x{size}' for size in filter_sizes]
    plot_results(input_image, filtered_images, plot_titles)
```
