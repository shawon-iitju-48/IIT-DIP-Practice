```python
from skimage import io
from skimage.transform import rescale, resize, downscale_local_mean
import matplotlib.pyplot as plt
import cv2
```

```python
img=io.imread('images/ironman.png',as_gray=True)
plt.imshow(img,cmap='gray')
```

```python
rescaled_img=rescale(img,.245,
    mode='constant',
    cval=0.0,
    anti_aliasing=False,
    channel_axis=None)
plt.imshow(rescaled_img,cmap='gray')
print("Old Size",img.shape)
print("New size: ",rescaled_img.shape)
```

```python
resized_image=resize(img,
    (92,160),
    mode='constant',
    anti_aliasing=None,)
plt.imshow(resized_image,cmap='gray')
```

```python
downscaled_img=downscale_local_mean(img, (3,3), cval=0.0)
print(downscaled_img.shape)
plt.imshow(downscaled_img,cmap='gray')
```

```python
image=cv2.imread('images/salty_e4.jpg',0)
plt.imshow(image,cmap='gray')
```

```python
resized=cv2.resize(image, None, fx=2,fy=2, interpolation=cv2.INTER_LINEAR)
plt.imshow(resized,cmap='gray')
```

```python
image1=cv2.imread('images/1.jpg',0)
image2=cv2.imread('images/2.jpg',0)
image2=cv2.resize(image2,(image1.shape[1],image1.shape[0]))
merged=cv2.addWeighted(image1, .6, image2, .4, 0)
plt.imshow(merged,cmap='gray')
```

```python
edged=cv2.Canny(image, 500, 800)
plt.imshow(edged,cmap='gray')
```
