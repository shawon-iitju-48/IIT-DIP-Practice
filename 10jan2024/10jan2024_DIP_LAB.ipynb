```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot  as plt
```

```python
print(cv2.__version__)
```

```python
img = cv2.imread("ironman.jpg",cv2.IMREAD_COLOR)
img = cv2.resize(img, (600, 300))  
cv2.imshow("& I am IronMan",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
color=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(color)
plt.title("& I am IronMan")
```

```python
img2 = cv2.imread("scarletwitch.jpg",cv2.IMREAD_COLOR)
img2 = cv2.resize(img2, (400, 700))  
cv2.imshow("A great mommy",img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
color2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(color2)
plt.title("A great Mommy")
```
