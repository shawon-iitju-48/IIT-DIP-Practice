# Answer-02

```python
from skimage import io
from skimage.filters import unsharp_mask
import matplotlib.pyplot as plt
from skimage.restoration import denoise_tv_chambolle

img=io.imread('Images/'+all_files[2],as_gray=True)

masked_img=unsharp_mask(img,
    radius=3.0,
    amount=10.5,
    channel_axis=None,)

tv_img=denoise_tv_chambolle(img,
    weight=0.3,
    eps=0.0002,
    max_num_iter=200,
    channel_axis=None,)

plt.subplot(2,2,1)
plt.title('Original Image')
plt.imshow(img,cmap='gray')

plt.subplot(2,2,2)
plt.title('Unsharp masked image')
plt.imshow(masked_img,cmap='gray')

plt.subplot(2,2,3)
plt.title('Original Image')
plt.imshow(img,cmap='gray')

plt.subplot(2,2,4)
plt.title('after total variance filtering')
plt.imshow(tv_img,cmap='gray')
```
