```python
import cv2
arr=[]
path="*.jpg"
 
for i in glob.glob(path):
    print(i)
    a=cv2.imread(i)
    a=cv2.cvtColor(a,cv2.COLOR_BGR2RGB)
    arr.append(a)
   
from matplotlib import pyplot as plt
plt.imshow(arr[0])
 
```

```python
path="*.jpg"
image_num=2

for file in glob.glob(path):
    img1=cv2.imread(file)
    c_img= cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
    cv2.imwrite("color-image"+str(image_num)+".jpg",c_img)
    plt.subplot(1, 3, image_num)
    image_num+=1
    plt.title(file)
    plt.imshow(c_img)
```
