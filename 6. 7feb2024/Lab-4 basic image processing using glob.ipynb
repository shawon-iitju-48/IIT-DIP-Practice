```python
path="*.jpg"
image_num=2

for file in glob.glob(path):
    img1=cv2.imread(file)
    c_img= cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
    cv2.imwrite("color-image"+str(image_num)+".jpg",c_img)
    plt.subplot(1, 3, image_num)
    image_num+=1
    plt.title(file)
    plt.imshow(c_img)
```

# Image Processing using scikit-image library

```python
from skimage import data, io, color, transform
from skimage.transform import rescale, resize , downscale_local_mean
```

```python
img=io.imread('man-of-culture.jpg',as_gray=True)
plt.imshow(img)
```

```python
img_rescaled=rescale(img,1.0/4.0,anti_aliasing=False)
plt.imshow(img_rescaled)
```

```python
img_resized=resize(img,(200,200),anti_aliasing=True)
plt.imshow(img_resized)
```

```python
img_downscaled=downscale_local_mean(img,(4,3))
plt.imshow(img_downscaled)
```
