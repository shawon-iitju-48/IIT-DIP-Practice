```python
from skimage import io, img_as_float
from skimage.filters import gaussian, sobel
from matplotlib import pyplot as plt
import cv2

img4 = cv2.imread("Images/ironman.png", 0)
fig = plt.figure(figsize=(6,6))

#applying gaussian filtering
gaussian_img=cv2.GaussianBlur(img4, (3,3), 0, borderType=cv2.BORDER_CONSTANT)

#applying edge detection algorithms
sobel_edge = sobel(gaussian_img)
canny_edge = cv2.Canny(gaussian_img, 100, 200)

#plotting the images
ax1 = fig.add_subplot(221)
ax1.imshow(img4, cmap="gray")
ax1.set_title('Input Image')
ax2 = fig.add_subplot(222)
ax2.imshow(gaussian_img, cmap="gray")
ax2.set_title('Filtered Image')
ax3 = fig.add_subplot(223)
ax3.imshow(sobel_edge, cmap='gray')
ax3.set_title('Edge detected using Sobel')
ax4 = fig.add_subplot(224)
ax4.imshow(canny_edge, cmap='gray')
ax4.set_title('Edge detected using Canny')
```

```python
import cv2
from skimage.restoration import denoise_bilateral
from skimage.restoration import denoise_tv_chambolle
from matplotlib import pyplot as plt

img = cv2.imread("Images/ironman.png", 0)
img2 = cv2.imread("Images/spiderman.png", 0)

fig = plt.figure(figsize=(8,4))

#bilateral
bilateral = cv2.bilateralFilter(img, 5, 20, 100, borderType=cv2.BORDER_CONSTANT)
bilateral2 = cv2.bilateralFilter(img2, 5, 20, 100, borderType=cv2.BORDER_CONSTANT)

#denoise_tv_chambolle
tv_filtered = denoise_tv_chambolle(img, weight=0.1, eps=0.0002, max_num_iter=200, channel_axis=False)
tv_filtered2 = denoise_tv_chambolle(img2, weight=0.1, eps=0.0002, max_num_iter=200, channel_axis=False)

#plotting for first input image
ax1 = fig.add_subplot(231)
ax1.imshow(img, cmap="gray")
ax1.set_title('Input Image 1')
ax2 = fig.add_subplot(232)
ax2.imshow(bilateral, cmap="gray")
ax2.set_title('Bilateral')
ax3 = fig.add_subplot(233)
ax3.imshow(tv_filtered, cmap='gray')
ax3.set_title('TV Chambolle')

#plotting for second input image
ax4 = fig.add_subplot(234)
ax4.imshow(img2, cmap="gray")
ax4.set_title('Input Image 2')
ax5 = fig.add_subplot(235)
ax5.imshow(bilateral2, cmap="gray")
ax5.set_title('Bilateral')
ax6 = fig.add_subplot(236)
ax6.imshow(tv_filtered2, cmap='gray')
ax6.set_title('TV Chambolle')
```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

img = cv2.imread("Images/ironman.png", 0)
img2 = cv2.imread("Images/spiderman.png", 0)
fig = plt.figure(figsize=(10,6))

#applying FT on first image
dft = cv2.dft(np.float32(img), flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift = np.fft.fftshift(dft)
magnitude_spectrum = 20 * np.log((cv2.magnitude(dft_shift[:, :, 0], dft_shift[:, :,1]))+1)

#applying FT on second image
dft2 = cv2.dft(np.float32(img2), flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift2 = np.fft.fftshift(dft2)
magnitude_spectrum2 = 20 * np.log((cv2.magnitude(dft_shift2[:, :, 0], dft_shift2[:, :,1]))+1)

ax1 = fig.add_subplot(221)
ax1.imshow(img)
ax1.set_title('Input Image-1')
ax2 = fig.add_subplot(222)
ax2.imshow(magnitude_spectrum)
ax2.set_title('FFT of Image-1')

ax3 = fig.add_subplot(223)
ax3.imshow(img2)
ax3.set_title('Input Image-2')
ax4 = fig.add_subplot(224)
ax4.imshow(magnitude_spectrum2)
ax4.set_title('FFT of Image-2')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("Images/spiderman.png", 0)

dft=cv2.dft(np.float32(img),flags=cv2.DFT_COMPLEX_OUTPUT)
dft_shift=np.fft.fftshift(dft)
magnitude_spectrum=20*np.log((cv2.magnitude(dft_shift[:,:,0],dft_shift[:,:,1]))+1)
#plotting input and fft 
fig=plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(4,2,1)
ax1.imshow(img, cmap='gray')
ax1.title.set_text('Input Image')
ax2 = fig.add_subplot(4,2,2)
ax2.imshow(magnitude_spectrum,cmap='gray')
ax2.title.set_text('FFT of image')

# High Pass Filter
rows,cols=img.shape
crow,ccol=int(rows/2),int(cols/2)
mask=np.ones((rows, cols, 2),np.uint8)
r=80
center=[crow, ccol]
x,y=np.ogrid[:rows, :cols]
mask_area=(x-center[0])**2+(y-center[1])**2<=r*r
mask[mask_area]=0
fshift=dft_shift*mask
epsilon = 1e-8
fshift_mask_mag=20*np.log(cv2.magnitude(fshift[:,:,0],fshift[:, :, 1])+epsilon)
f_ishift=np.fft.ifftshift(fshift)
img_back=cv2.idft(f_ishift)
img_back=cv2.magnitude(img_back[:,:,0],img_back[:,:,1])
#plotting
fig=plt.figure(figsize=(12,12))
ax3=fig.add_subplot(4,2,3)
ax3.imshow(fshift_mask_mag, cmap='gray')
ax3.title.set_text('FFT + high pass filter Mask')
ax4 = fig.add_subplot(4,2,4)
ax4.imshow(img_back, cmap='gray')
ax4.title.set_text('After inverse FFT usign high pass filter')
plt.show()


# Low pass filter
rows, cols = img.shape
crow, ccol = int(rows / 2), int(cols / 2)
mask = np.zeros((rows, cols, 2), np.uint8)
r = 100
center = [crow, ccol]
x, y = np.ogrid[:rows, :cols]
mask_area = (x - center[0]) ** 2 + (y - center[1]) ** 2 <= r*r
mask[mask_area] = 1
fshift = dft_shift * mask
epsilon = 1e-8
fshift_mask_mag = 20 * np.log(cv2.magnitude(fshift[:, :, 0], fshift[:, :, 1]) + epsilon)
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:, :, 0], img_back[:, :, 1])
#plotting
fig = plt.figure(figsize=(12,12))
ax3 = fig.add_subplot(4,2,5)
ax3.imshow(fshift_mask_mag, cmap='gray')
ax3.title.set_text('FFT + low pass filter Mask')
ax4 = fig.add_subplot(4,2,6)
ax4.imshow(img_back, cmap='gray')
ax4.title.set_text('After inverse FFT using low pass filter')
plt.show()

# Band Pass Filter
rows, cols = img.shape
crow, ccol = int(rows / 2), int(cols / 2)
mask = np.zeros((rows, cols, 2), np.uint8)
r_out = 80
r_in = 10
center = [crow, ccol]
x, y = np.ogrid[:rows, :cols]
mask_area = np.logical_and(((x - center[0]) ** 2 + (y - center[1]) ** 2 >= r_in ** 2),
((x - center[0]) ** 2 + (y - center[1]) ** 2 <= r_out ** 2))
mask[mask_area] = 1
fshift = dft_shift * mask
epsilon = 1e-8
fshift_mask_mag = 20 * np.log(cv2.magnitude(fshift[:, :, 0], fshift[:, :, 1]) + epsilon)
f_ishift = np.fft.ifftshift(fshift)
img_back = cv2.idft(f_ishift)
img_back = cv2.magnitude(img_back[:, :, 0], img_back[:, :, 1])
#plotting
fig = plt.figure(figsize=(12,12))
ax3 = fig.add_subplot(4,2,7)
ax3.imshow(fshift_mask_mag, cmap='gray')
ax3.title.set_text('FFT +band pass filter Mask')
ax4 = fig.add_subplot(4,2,8)
ax4.imshow(img_back, cmap='gray')
ax4.title.set_text('After inverse FFT using band pass filter')
plt.show()
```
